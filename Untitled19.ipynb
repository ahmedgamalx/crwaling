```python

def extract_urls(text):

  urls = []
  for match in re.finditer(r'(https:\/\/www\.|http:\/\/www\.|https:\/\/|http:\/|http:\/)+[a-zA-Z0-9./]*(\.[a-zA-Z0-9/-]*)(\.[a-zA-Z0-9/:_%.-a-zA-Z0-9?]{2,})+', text):
    url = match.group()
    urls.append(url)
  return urls




```

```python
urls = extract_urls(content)

for url in urls:
  print(url)
```

```python
from urllib.parse import urlparse

def filter_urls_by_domain(urls, domain):
    filtered_urls = []

    for url in urls:
        parsed_url = urlparse(url)
        if parsed_url.netloc == domain:
            filtered_urls.append(url)

    return filtered_urls
```

```python
def classify_url2(url):
    image_extensions = ["jpg", "jpeg", "png", "gif", "bmp", "tiff", "webp"]
    document_extensions = ["pdf", "doc", "docx", "xls", "xlsx", "ppt", "pptx"]
    video_extensions = ["mp4", "avi", "mov", "mkv", "flv", "mpeg", "mpg"]
    audio_extensions = ["mp3", "wav", "aiff", "ogg", "flac"]

    extension = url.split(".")[-1].lower()

    if extension in image_extensions:
        return "Image"
    elif extension in document_extensions:
        return "Document"
    elif extension in video_extensions:
        return "Video"
    elif extension in audio_extensions:
        return "Audio"
    else:
        return "site"

```

```python
def count_repetition(list1):
  counts = {}
  for item in list1:
    if item in counts:
      counts[item] += 1
    else:
      counts[item] = 1
  return counts
```
